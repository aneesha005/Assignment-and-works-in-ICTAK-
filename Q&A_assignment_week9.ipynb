**3.The manager wants to find out whether the same amount was spent for the three advertisements(TV,Radio and Newspaper). comment on your findings **

H0: The amount spend on the three advertisements are same.

H1: The amount spend on the three advertisements are not same.

```python
import scipy.stats as stats
```

```python
fvalue,pvalue=stats.f_oneway(df['TV'],df['Newspaper'],df['Radio'])
print('F-statistics=',fvalue,'Pvalue=',pvalue)
```

The pvalue is less than 0.05 we can REJECT the Null Hypothesis(H0),The amount spend on the three advertisements are not same.
